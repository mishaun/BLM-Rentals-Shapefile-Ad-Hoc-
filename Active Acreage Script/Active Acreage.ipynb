# Reading In Rental List - Magnum and R&R

```python
#Reading Mastr shapefile
shapefile = gpd.read_file("zip://BLM Master 7-2021.zip")
shapefile.columns
```

# Filter Expired Acreage

```python
todaysDate = dt.datetime.now()

todaysDate = todaysDate.strftime('%Y-%m-%d')
```

```python
todaysDate
```

```python
activeAcreageShape = shapefile.loc[shapefile['EXPIRATION']>todaysDate]

manualDate = '2021-08-1'
TestactiveAcreageShape = shapefile.loc[shapefile['EXPIRATION']>manualDate]
```

### Looking at lenghts of filtered shapefiles to make sure it worked

```python
print("Unfiltered length of data set: {}".format(shapefile.shape[0]))
print("Filtered by today's date length of data set: {}".format(activeAcreageShape.shape[0]))
print("Filtered by manual date entered, length of data set: {}".format(TestactiveAcreageShape.shape[0]))
```

```python
TestactiveAcreageShape.to_file("ActiveAcreage as of {}.shp".format(manualDate))
```
